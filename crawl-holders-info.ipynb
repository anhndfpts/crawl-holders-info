```python
total_ticker_lst = delisted_ticker_lst + ticker_lst
total_ticker_lst = list(set(total_ticker_lst))
```

```python
from tqdm import tqdm
from threading import Thread
import time
import random

lst_1 = total_ticker_lst[0:400]
lst_2 = total_ticker_lst[400:800]
lst_3 = total_ticker_lst[800:1200]
lst_4 = total_ticker_lst[1200:1600]
lst_5 = total_ticker_lst[1600:]

lst_threads_ticker = [lst_1, lst_2, lst_3, lst_4, lst_5]

def crawl_ticker_data(*args):
    lst_ticker = list(args)
    while len(lst_ticker) > 0:
        try:
            df_temp = crawl_holders_info(lst_ticker[0])
            df_temp.to_pickle(rf"D:/fpa_product/data/holders_info/{lst_ticker[0]}.pkl")
            df_temp = crawl_holders_transactions(lst_ticker[0], limit=1000)
            df_temp.to_pickle(rf"D:/fpa_product/data/holders_transactions/{lst_ticker[0]}.pkl")
            print(lst_ticker[0])
        except Exception as e:
            print(e)
            time.sleep(random.randint(20,50))
            continue
        lst_ticker.remove(lst_ticker[0])

threads = [Thread(target=crawl_ticker_data, args=lst_ticker) for lst_ticker in lst_threads_ticker]

# start the threads
for thread in threads:
    thread.start()

# wait for the threads to complete
for thread in threads:
    thread.join()

# for ticker in tqdm(total_ticker_lst):
#     # print(ticker)
#     df = crawl_holders_info(ticker)
#     df.to_pickle(rf"D:/fpa_product/data/holders_info/{ticker}.pkl")
#     df = crawl_holders_transactions(ticker, limit=1000)
#     df.to_pickle(rf"D:/fpa_product/data/holders_transactions/{ticker}.pkl")
```
